```python
preds = {}

for preds_name in os.listdir('../data/preds'):
    if preds_name.endswith('.npy'):
#         print(preds_name)
        
        with open( f'../data/preds/{preds_name}', 'rb') as f:
            
#             print(f)
            posts   = np.load(f)
            targets = np.load(f)
            inds    = np.load(f)
            obs0s   = np.load(f)
            
        pred = dict(posts = posts, targets = targets, inds = inds, obs0s = obs0s)
        
        preds[preds_name] = pred
        
```

```python
def get_title(title, m_titles):
    titles = np.array(np.copy(m_titles), dtype = object)
    titles[0] = title + "\n" +  m_titles[0]
    return titles
```

```python
def f(posts, targets, x = np.linspace(0, 1)):
    
    x1 = x[:-1]
    x2 = x[1:]
    x3 = np.mean([x1, x2], axis = 0)
    
    def f1(x):
        y = targets[posts > x]
        return y.sum() / len(y) if y.sum() != 0 else 0

    def f2(x):
        y = targets[posts < x]
        return y.sum() / len(y) if y.sum() != 0 else 0

    def f3(x1, x2):
        y = targets[(posts > x1) & (posts < x2)]
        return y.sum() / len(y) if y.sum() != 0 else 0
    
    y1 = np.array([f1(xi) for xi in x1])
#     y2 = np.array([f2(xi) for xi in x2])
    y3 = np.array([f3(xi, xj) for xi, xj in zip(x1, x2)])


    return [x1, x3], [y1, y3]
    
```
