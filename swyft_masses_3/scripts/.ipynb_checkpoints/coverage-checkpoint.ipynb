```python
preds = {}

for preds_name in os.listdir('../data/preds'):
    if preds_name.endswith('.npy'):
#         print(preds_name)
        
        with open( f'../data/preds/{preds_name}', 'rb') as f:
            
#             print(f)
            posts   = np.load(f)
            targets = np.load(f)
            inds    = np.load(f)
            obs0s   = np.load(f)
            
        pred = dict(posts = posts, targets = targets, inds = inds, obs0s = obs0s)
        
        preds[preds_name] = pred
        
```

```python
def get_title(title, m_titles):
    titles = np.array(np.copy(m_titles), dtype = object)
    titles[0] = title + "\n" +  m_titles[0]
    return titles
```

```python
keys = list(preds.keys())
keys = [keys[-1]]
keys
```

```python
posts.shape = (10_000, 1, 40, 40)
```

```python
def f(posts, targets, x):
    
    x1, x2 = x[:-1], x[1:]
    
    def f3(xi, xj):
        y = targets[(posts > xi) & (posts < xj)]
        return y.sum() / len(y) if y.sum() != 0 else 0

    return np.array([f3(xi, xj) for xi, xj in zip(x1, x2)])
```

```python
alpha_edges = np.linspace(0, 1, 100)
alpha_centers = (alpha_edges[:-1] + alpha_edges[1:])/2
```

```python
posts_all = ([])
targets_all = ([])
fs = ([])
obs0s_all = ([])


for key in keys:

    name = key
    posts   = preds[key]['posts']
    targets = preds[key]['targets']
    inds    = preds[key]['inds']
    obs0s   = preds[key]['obs0s']
    
    obs0s = np.expand_dims(obs0s, 1)
    
    
    
    
    _, nsub, nsim, nmbins, _, _, _, _, _, nobs = list(map(int, re.findall(r'[0-9]+', key)))
    print('##### nsub {}, nsim {}, nmbins {}, nobs {} ####'.format(nsub, nsim, nmbins, nobs))
    
    
    m = np.logspace(10, 12, 2*nmbins+1)
    m_centers, m_edges = m[1::2], m[0::2]
    m_titles = [rf'$M_{{sub}} = 10^{ {np.round(m, 2)} }\ M_{{\odot}}$' for m in np.log10(m_centers)]
    m_titles = [rf'$M_{{sub}} = 10^{ {np.round(m1, 2)} } - 10^{ {np.round(m2, 2)} }\ M_{{\odot}}$' for m1, m2 in zip(np.log10(m_edges)[:-1], np.log10(m_edges)[1:])]
    
    for i in np.random.randint(0, nobs, 1):
        target_coords = np.array(np.where(targets[i] == 1)).T
        plt_imshow(posts[i], target_coords = target_coords, 
                   titles = get_title('Prediction  ' + rf'$N_{{sub}} = {nsub}, N_{{\mathrm{{m bins}}}} = {nmbins}$', m_titles),
                   title_size = 15,
                   y = 4, cbar = True, tl = True)
        
        plt_imshow(np.concatenate((obs0s[i], targets[i], 1 - targets[i])),
                   target_coords = target_coords, 
                   titles = np.concatenate((
                       get_title('Observation  ' + rf'$N_{{sub}} = {nsub}, N_{{\mathrm{{m bins}}}} = {nmbins}$', m_titles),
                       np.array(('Target halo', 'Target no halo'))
                    )),
                   title_size = 15,
                   y = 4, cbar = True, tl = True)
    
    plt_imshow(
        obs0s[i:i+4].squeeze(),
        titles = ['Observation \n obs 1', 'obs 2', 'obs 3', 'obs 4'],
        title_size = 15,
        cbar = True,
    )
    plt_imshow(
        posts[i:i+4].squeeze(),
        titles = ['Posterior probabilites \n obs 1', 'obs 2', 'obs 3', 'obs 4'],
        title_size = 15,
        cbar = True,
    )
        
    posts_all.append(posts)
    targets_all.append(targets)
    obs0s_all.append(obs0s)
    fs.append(f(posts, targets, alpha_edges))

fs = np.array(fs)
```
